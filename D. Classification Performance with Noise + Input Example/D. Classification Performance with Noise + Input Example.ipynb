```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import euclidean_distances
import tensorflow as tf
from tensorflow.keras import layers
import re
from matplotlib.backends.backend_pdf import PdfPages
import traceback
from scipy import stats
from sklearn.preprocessing import RobustScaler

# 1. DATA LOADING FUNCTION
def load_data(folder, interp_points=1000):
    spectra, frequencies, headers, filenames_list, logn_values, tex_values = [], [], [], [], [], []
    
    for filename in os.listdir(folder):
        if "_simulate_generate" in filename and "_cleaned_simulate_generate" not in filename:
            filepath = os.path.join(folder, filename)
            with open(filepath, 'r', encoding='utf-8') as file:
                lines = file.readlines()
                if len(lines) < 3:
                    continue
                
                header = lines[0].strip()
                logn_match = re.search(r'logn[\s=:]+([0-9]+\.?[0-9]*)', header.lower())
                tex_match = re.search(r'tex[\s=:]+([0-9]+\.?[0-9]*)', header.lower())
                
                if logn_match and tex_match:
                    try:
                        logn_value = float(logn_match.group(1))
                        tex_value = float(tex_match.group(1))
                        if not (12 <= logn_value <= 18 and 10 <= tex_value <= 500):
                            continue
                    except (ValueError, TypeError):
                        continue
                else:
                    continue
                
                data_lines = [line for line in lines[2:] if line.strip() and not line.startswith("//")]
                try:
                    data = np.array([list(map(float, line.split())) for line in data_lines if len(line.split()) >= 2])
                    if len(data) < 10:
                        continue
                    freq, spec = data[:, 0], data[:, 1]
                    frequencies.append(freq)
                    spectra.append(spec)
                    headers.append(header)
                    filenames_list.append(filename)
                    logn_values.append(logn_value)
                    tex_values.append(tex_value)
                except (ValueError, IndexError):
                    continue
    
    if not spectra:
        raise ValueError(f"No valid data found in {folder}")
    
    min_freq = max([np.min(freq) for freq in frequencies])
    max_freq = min([np.max(freq) for freq in frequencies])
    if min_freq >= max_freq:
        raise ValueError("Invalid frequency range in spectra")
    
    new_x = np.linspace(min_freq, max_freq, interp_points)
    
    interpolated_spectra = []
    scaler = RobustScaler(quantile_range=(5, 95))
    
    for freq, spec in zip(frequencies, spectra):
        try:
            unique_freq, unique_idx = np.unique(freq, return_index=True)
            unique_spec = spec[unique_idx]
            interp_spec = np.interp(new_x, unique_freq, unique_spec, left=0, right=0)
            scaled_spec = scaler.fit_transform(interp_spec.reshape(-1, 1)).flatten()
            interpolated_spectra.append(scaled_spec)
        except:
            continue
    
    if not interpolated_spectra:
        raise ValueError("No valid spectra after processing")
    
    print(f"Loaded {len(interpolated_spectra)} spectra from {folder}")
    return (np.tile(new_x, (len(interpolated_spectra), 1)), 
            np.array(interpolated_spectra), 
            headers, filenames_list, 
            np.array(logn_values), np.array(tex_values))

# 2. CNN MODEL BUILDING
def build_cnn_model(input_shape):
    model = tf.keras.Sequential([
        tf.keras.Input(shape=input_shape),
        layers.Conv1D(64, kernel_size=7, activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Conv1D(128, kernel_size=5, activation='relu', padding='same'),
        layers.MaxPooling1D(2, padding='same'),
        layers.Dropout(0.3),
        layers.Conv1D(256, kernel_size=3, activation='relu', padding='same'),
        layers.GlobalAveragePooling1D(),
        layers.Dense(128, activation='relu'),
        layers.Dense(input_shape[0], activation='linear')
    ])
    
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                 loss='mse',
                 metrics=['mae'])
    return model

# 3. MODEL TRAINING
def train_and_evaluate_cnn(model, train_data):
    callbacks = [
        tf.keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True),
        tf.keras.callbacks.ReduceLROnPlateau(factor=0.5, patience=5)
    ]
    
    history = model.fit(
        train_data, train_data,
        epochs=100,
        batch_size=32,
        validation_split=0.15,
        callbacks=callbacks,
        verbose=1
    )
    
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Train Loss', alpha=0.5)
    plt.plot(history.history['val_loss'], label='Val Loss', alpha=0.5)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['mae'], label='Train MAE', alpha=0.5)
    plt.plot(history.history['val_mae'], label='Val MAE', alpha=0.5)
    plt.xlabel('Epoch')
    plt.ylabel('MAE')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    
    return model

# 4.PLOTTING FUNCTIONS WITH REQUESTED MODIFICATIONS
def plot_logn_vs_predicted(test_logn, train_logn, closest_matches, test_tex, database_folder, pdf):
    fig, ax = plt.subplots(figsize=(10, 8))
    
    pred_logn = train_logn[closest_matches]
    
    sc = ax.scatter(test_logn, pred_logn, c=test_tex, cmap='viridis',
                   alpha=0.5, edgecolors='k', vmin=np.percentile(test_tex, 5), vmax=np.percentile(test_tex, 95))
    plt.colorbar(sc, label='Tex (K)')
    
    # Plot ideal line
    min_val = min(np.min(test_logn), np.min(pred_logn))
    max_val = max(np.max(test_logn), np.max(pred_logn))
    ax.plot([min_val, max_val], [min_val, max_val], 'r--', label='Ideal', alpha=0.5)
    
    ax.set_xlabel('True LogN (cm⁻²)')
    ax.set_ylabel('Predicted LogN (cm⁻²)')
    ax.set_title(f'LogN Comparison - {os.path.basename(database_folder)}')
    ax.grid(alpha=0.3)
    ax.legend()
    
    pdf.savefig(fig, bbox_inches='tight')
    plt.close(fig)

def plot_tex_vs_predicted(test_tex, train_tex, closest_matches, test_logn, database_folder, pdf):
    fig, ax = plt.subplots(figsize=(10, 8))
    
    pred_tex = train_tex[closest_matches]
    
    sc = ax.scatter(test_tex, pred_tex, c=test_logn, cmap='plasma',
                   alpha=0.5, edgecolors='k', vmin=np.percentile(test_logn, 5), vmax=np.percentile(test_logn, 95))
    plt.colorbar(sc, label='LogN (cm⁻²)')
    
    # Plot ideal line with exact 45 degree slope
    min_val = min(np.min(test_tex), np.min(pred_tex))
    max_val = max(np.max(test_tex), np.max(pred_tex))
    ax.plot([min_val, max_val], [min_val, max_val], 'r--', label='Ideal (45°)', alpha=0.5)
    
    ax.set_xlabel('True Tex (K)')
    ax.set_ylabel('Predicted Tex (K)')
    ax.set_title(f'Tex Comparison - {os.path.basename(database_folder)}')
    ax.grid(alpha=0.3)
    ax.legend()
    
    ax.set_aspect('equal', adjustable='box')
    
    pdf.savefig(fig, bbox_inches='tight')
    plt.close(fig)

def plot_spectra(test_data, test_freq, train_data, train_freq, closest_matches, 
                test_filenames, test_headers, train_filenames, train_headers, 
                reconstructed_test, pdf):
    n_samples = min(10, len(test_data))
    indices = np.random.choice(len(test_data), n_samples, replace=False)
    
    fig = plt.figure(figsize=(15, 3*n_samples))
    for i, idx in enumerate(indices):
        ax = fig.add_subplot(n_samples, 1, i+1)
        best_match_idx = closest_matches[idx]
        
        ax.plot(test_freq[idx], test_data[idx].flatten(), 
               label='Test Spectrum', color='blue', linewidth=1.5, alpha=0.5)
        ax.plot(train_freq[best_match_idx], train_data[best_match_idx].flatten(), 
               label='Best Match', linestyle='--', color='green', linewidth=1.5, alpha=0.5)
        ax.plot(test_freq[idx], reconstructed_test[idx].flatten(), 
               label='CNN Reconstruction', linestyle=':', color='red', linewidth=2, alpha=0.5)
        
        test_params = re.search(r'logn=([\d.]+).*tex=([\d.]+)', test_headers[idx].lower())
        match_params = re.search(r'logn=([\d.]+).*tex=([\d.]+)', train_headers[best_match_idx].lower())
        
        info_text = (f"Test: LogN={test_params.group(1) if test_params else '?'} "
                    f"Tex={test_params.group(2) if test_params else '?'}\n"
                    f"Match: LogN={match_params.group(1) if match_params else '?'} "
                    f"Tex={match_params.group(2) if match_params else '?'}")
        
        ax.text(0.02, 0.95, info_text, transform=ax.transAxes,
               fontsize=8, verticalalignment='top', bbox=dict(facecolor='white', alpha=0.7))
        
        ax.legend(loc='upper right', fontsize=8)
        ax.set_xlabel("Frequency (Hz)", fontsize=9)
        ax.set_ylabel("Intensity", fontsize=9)
    
    fig.suptitle("Spectra Comparison: Test vs Best Matches", y=1.02, fontsize=12)
    fig.tight_layout()
    pdf.savefig(fig)
    plt.close(fig)

#5.INPUT SPECTRUM PROCESSING WITH IMPROVED VISUALIZATION
def process_input_spectrum(filepath, model, train_data, encoded_train, 
                         train_freq, train_filenames, train_headers, pdf):
    try:
        print(f"\nProcessing input spectrum: {filepath}")
        
        if not os.path.exists(filepath):
            raise FileNotFoundError(f"File not found: {filepath}")
        
        with open(filepath, 'r', encoding='utf-8') as file:
            lines = file.readlines()
            header = lines[0].strip() if lines else ""
            
            data = []
            for line in lines[1:]:
                line = line.strip()
                if line and not line.startswith(("//", "#")):
                    parts = re.split(r'[\s,;]+', line)
                    if len(parts) >= 2:
                        try:
                            freq = float(parts[0]) * 1e9  # Convert GHz to Hz
                            intensity = float(parts[1])
                            data.append((freq, intensity))
                        except ValueError:
                            continue
            
            if len(data) < 10:
                raise ValueError("Insufficient valid data points in spectrum")
            
            freq, spec = zip(*data)
            freq = np.array(freq)
            spec = np.array(spec)
        
        train_min_freq = np.min(train_freq)
        train_max_freq = np.max(train_freq)
        new_x = np.linspace(train_min_freq, train_max_freq, train_data.shape[1])
        
        valid_mask = (freq >= train_min_freq) & (freq <= train_max_freq)
        if not np.any(valid_mask):
            raise ValueError("Input spectrum frequency range does not overlap with training data")
        
        interp_spec = np.interp(new_x, freq[valid_mask], spec[valid_mask], left=0, right=0)
        
        scaler = RobustScaler(quantile_range=(5, 95))
        normalized_spec = scaler.fit_transform(interp_spec.reshape(-1, 1)).flatten()
        
        input_spec = np.expand_dims(normalized_spec, axis=[0, -1])
        
        encoded_spec = model.predict(input_spec, verbose=0)
        reconstructed = model.predict(input_spec, verbose=0)
        
        if reconstructed.ndim == 3:
            reconstructed = reconstructed[0,:,0]  # Shape (1, N, 1) -> (N,)
        else:
            reconstructed = reconstructed[0]  # Shape (1, N) -> (N,)
        
        distances = euclidean_distances(encoded_spec.reshape(1, -1), 
                                      encoded_train.reshape(len(encoded_train), -1))
        closest_idx = np.argmin(distances)
        closest_dist = distances[0][closest_idx]
        
        fig = plt.figure(figsize=(16, 12))
        gs = plt.GridSpec(2, 1, height_ratios=[3, 1], hspace=0.3)
        
        ax = fig.add_subplot(gs[0])
        
        ax.plot(new_x, normalized_spec, label='Input Spectrum', color='navy', linewidth=2, alpha=0.5)
        ax.plot(train_freq[closest_idx], train_data[closest_idx,:,0], 
               label=f'Best Match (distance={closest_dist:.4f})', 
               linestyle='--', color='green', linewidth=2, alpha=0.5)
        ax.plot(new_x, reconstructed, 
               label='CNN Reconstruction', 
               linestyle=':', color='red', linewidth=2.5, alpha=0.5)
        
        all_data = np.concatenate([normalized_spec, 
                                 train_data[closest_idx,:,0].flatten(), 
                                 reconstructed])
        y_min = np.min(all_data) - 0.1 * np.ptp(all_data)
        y_max = np.max(all_data) + 0.1 * np.ptp(all_data)
        ax.set_ylim(y_min, y_max)
        
        ax.set_xlabel("Frequency (Hz)", fontsize=12)
        ax.set_ylabel("Normalized Intensity", fontsize=12)
        ax.set_title(f"Input Spectrum Analysis: {os.path.basename(filepath)}", fontsize=14)
        ax.grid(alpha=0.2)
        ax.legend(fontsize=10, loc='upper right')
        
        text_ax = fig.add_subplot(gs[1])
        text_ax.axis('off')
        
        input_params = re.search(r'logn=([\d.]+).*tex=([\d.]+)', header.lower()) if header else None
        match_params = re.search(r'logn=([\d.]+).*tex=([\d.]+)', train_headers[closest_idx].lower())
        
        info_text = (
            f"INPUT SPECTRUM:\n"
            f"{'='*50}\n"
            f"• File: {os.path.basename(filepath)}\n"
            f"• LogN: {input_params.group(1) if input_params else 'Not specified'}\n"
            f"• Tex: {input_params.group(2) if input_params else 'Not specified'}\n\n"
            f"BEST MATCH IN DATABASE:\n"
            f"{'='*50}\n"
            f"• File: {train_filenames[closest_idx]}\n"
            f"• Distance: {closest_dist:.6f}\n"
            f"• LogN: {match_params.group(1) if match_params else 'Unknown'}\n"
            f"• Tex: {match_params.group(2) if match_params else 'Unknown'}\n"
            f"• Header: {train_headers[closest_idx][:200]}..."
        )
        
        text_ax.text(0.5, 0.5, info_text, ha='center', va='center', fontsize=10,
                   bbox=dict(facecolor='whitesmoke', alpha=0.7, boxstyle='round'),
                   fontfamily='monospace', linespacing=1.8)
        
        plt.tight_layout()
        pdf.savefig(fig, bbox_inches='tight')
        plt.close(fig)
        
    except Exception as e:
        print(f"Error processing input spectrum: {str(e)}")
        traceback.print_exc()

# 6. MAIN EXECUTION
def main():
    directories = [
        r"C:\1. AI - ITACA\9. JupyterNotebook\_BK\SYNTHETIC_SPECTRA_NOISE_6\JPL_CH3CN",
        r"C:\1. AI - ITACA\9. JupyterNotebook\_BK\SYNTHETIC_SPECTRA_NOISE_6\JPL_CO",
        r"C:\1. AI - ITACA\9. JupyterNotebook\_BK\SYNTHETIC_SPECTRA_NOISE_6\JPL_HCO+v=0,1,2",
        r"C:\1. AI - ITACA\9. JupyterNotebook\_BK\SYNTHETIC_SPECTRA_NOISE_6\JPL_SiO",
        r"C:\1. AI - ITACA\9. JupyterNotebook\_BK\SYNTHETIC_SPECTRA_NOISE_6\ALL"
    ]

    input_spectrum_file = r"C:\1. AI - ITACA\9. JupyterNotebook\_BK\example_4mols_format"

    for database_folder in directories:
        if not os.path.exists(database_folder):
            print(f"\nDirectory not found: {database_folder}")
            continue
            
        pdf_filename = f"CNN_PERFORMANCE_ANALYSIS_with_noise{os.path.basename(database_folder)}.pdf"
        
        with PdfPages(pdf_filename) as pdf:
            try:
                print(f"\nProcessing database: {database_folder}")
                
                frequencies, data, headers, filenames, logn_values, tex_values = load_data(database_folder)
                
                if len(data) < 20:
                    print(f"Not enough samples in {database_folder} ({len(data)}), skipping...")
                    continue
                
                n_bins = min(5, len(np.unique(logn_values)))
                bins = np.linspace(np.min(logn_values), np.max(logn_values), n_bins+1)
                y_binned = np.digitize(logn_values, bins)
                
                (train_data, test_data, 
                 train_freq, test_freq, 
                 train_headers, test_headers,
                 train_filenames, test_filenames,
                 train_logn, test_logn,
                 train_tex, test_tex) = train_test_split(
                    data, frequencies, headers, filenames, logn_values, tex_values,
                    test_size=0.2, random_state=42, stratify=y_binned)
                
                train_data = np.expand_dims(train_data, axis=-1)
                test_data = np.expand_dims(test_data, axis=-1)
                
                print("\nTraining CNN model...")
                model = build_cnn_model((train_data.shape[1], 1))
                model = train_and_evaluate_cnn(model, train_data)
                
                print("\nGenerating embeddings...")
                encoded_train = model.predict(train_data, verbose=0)
                encoded_test = model.predict(test_data, verbose=0)
                
                distances = euclidean_distances(encoded_test.reshape(len(encoded_test), -1),
                                          encoded_train.reshape(len(encoded_train), -1))
                closest_matches = np.argmin(distances, axis=1)
                
                print("\nGenerating analysis plots...")
                plot_logn_vs_predicted(test_logn, train_logn, closest_matches, test_tex, database_folder, pdf)
                plot_tex_vs_predicted(test_tex, train_tex, closest_matches, test_logn, database_folder, pdf)
                
                reconstructed_test = model.predict(test_data, verbose=0)
                plot_spectra(test_data, test_freq, train_data, train_freq, closest_matches,
                            test_filenames, test_headers, train_filenames, train_headers,
                            reconstructed_test, pdf)
                
                if os.path.exists(input_spectrum_file):
                    print("\nProcessing input spectrum...")
                    process_input_spectrum(input_spectrum_file, model, train_data, encoded_train,
                                         train_freq, train_filenames, train_headers, pdf)
                else:
                    print(f"\nInput spectrum not found at: {input_spectrum_file}")
                
                print(f"\nAnalysis complete. Results saved to: {pdf_filename}")
                
            except Exception as e:
                print(f"\nError processing {database_folder}: {str(e)}")
                traceback.print_exc()

if __name__ == "__main__":
    tf.keras.utils.set_random_seed(42)
    tf.config.experimental.enable_op_determinism()
    main()
```
